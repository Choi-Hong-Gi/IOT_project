```python
import csv
import matplotlib.pyplot as plt
import numpy as np

############################################# 데이터 추출/저장 #############################################
file1 = open('GDP.csv')
data1 = csv.reader(file1)
next(file1)
file2 = open('ER.csv')
data2 = csv.reader(file2)
next(file2)
file3 = open('SI.csv')
data3 = csv.reader(file3)
next(file3)

GDP = []
ER = []
SI = []
x=[]
x_1 = []

for year in range(1980,2022):
    x.append(year)
    
for year in range(1980,2021):
    x_1.append(year)

for row1 in data1:
    for i1 in range(len(row1)):
        GDP.append(float(row1[i1]))
        
for row2 in data2:
    for i2 in range(len(row2)):
        ER.append(float(row2[i2]))
        
for row3 in data3:
    for i3 in range(len(row3)):
        SI.append(float(row3[i3]))
        
# print(x, '\n')
# print(GDP, "\n")
# print(ER, "\n")
# print(SI, "\n")
############################################# 데이터 추출/저장 #############################################

############################################# 매핑 #############################################
def map(x,input_min,input_max,output_min,output_max):
    return (x - input_min) * (output_max - output_min) / (input_max - input_min)+output_min

GDP_map = []
ER_map = []
SI_map = []

for i in GDP:
    i = map(i,min(GDP), max(GDP), 1, 100)
    GDP_map.append(i)
    
for j in ER:
    j = map(j, min(ER), max(ER), 1, 100)
    ER_map.append(j)
    
for k in SI:
    k = map(k, min(SI), max(SI), 1, 100)
    SI_map.append(k)
        
# print(GDP_map)
# print(ER_map)
# print(SI_map)
############################################# 매핑 #############################################

############################################# 결과 #############################################
#---------------------------------------- GDP ----------------------------------------
max_differ1 = 0 #최대 성장
max_index1 = 0 #최대 성장 년도
min_differ1 = 100 #최소 성장
min_index1 = 0 #최소 성장 년도

GDP_CR = [] #GDP 변화량

for pre1 in range(len(GDP)-1):
    a = GDP[pre1+1]/GDP[pre1]
    GDP_CR.append(a)
    if a > max_differ1:
        max_differ1 = a
        max_index1 = pre1
    if a < min_differ1:
        min_differ1 = a
        min_index1 = pre1

print("GDP 최대 상승 해: ", max_index1+1980,"년, 전년대비",int(GDP_CR[max_index1]*100),"%",
      "\nGDP 최대 감소 해: ", min_index1+1980, "년, 전년대비",int(GDP_CR[min_index1]*100),"%")
#---------------------------------------- GDP ----------------------------------------

#---------------------------------------- ER ----------------------------------------
max_differ2 = 0 #최대 성장
max_index2 = 0 #최대 성장 년도
min_differ2 = 100 #최소 성장
min_index2 = 0 #최소 성장 년도

ER_CR = [] #ER 변화량

for pre2 in range(len(ER)-1):
    b = ER[pre2+1]/ER[pre2]
    ER_CR.append(b)
    if b > max_differ2:
        max_differ2 = b
        max_index2 = pre2
    if b < min_differ2:
        min_differ2 = b
        min_index2 = pre2

print("\n환율 최대 상승 해: ", max_index2+1980,"년, 전년대비",int(ER_CR[max_index2]*100),"%",
      "\n환율 최대 감소 해: ", min_index2+1980, "년, 전년대비",int(ER_CR[min_index2]*100),"%")
#---------------------------------------- ER ----------------------------------------

#---------------------------------------- SI ----------------------------------------
max_differ3 = 0 #최대 성장
max_index3 = 0 #최대 성장 년도
min_differ3 = 100 #최소 성장
min_index3 = 0 #최소 성장 년도

SI_CR = [] #SI 변화량

for pre3 in range(len(SI)-1):
    c = SI[pre3+1]/SI[pre3]
    SI_CR.append(c)
    if c > max_differ3:
        max_differ3 = c
        max_index3 = pre3
    if c < min_differ3:
        min_differ3 = c
        min_index3 = pre3

print("\n주가지수 최대 상승 해: ", max_index3+1980,"년, 전년대비",int(SI_CR[max_index3]*100),"%",
      "\n주가지수 최대 감소 해: ", min_index3+1980, "년, 전년대비",int(SI_CR[min_index3]*100),"%",)
#---------------------------------------- SI ----------------------------------------
############################################# 결과 #############################################
       
############################################# 그래프 #############################################
plt.figure(dpi = 75)
plt.title('GDP')
plt.plot(x, GDP,'orange')
#plt.grid(True)

# plt.figure(dpi = 75)
# plt.title('GDP Change')
# plt.plot(x_1, GDP_CR,'orange')

plt.figure(dpi = 75)
plt.title('Exchange Rate ($)')
plt.plot(x, ER, 'violet')

# plt.figure(dpi = 75)
# plt.title('Exchange Rate ($) Change')
# plt.plot(x_1, ER_CR, 'violet')

plt.figure(dpi = 75)
plt.title('Stock Index')
plt.plot(x, SI,'g')

# plt.figure(dpi = 75)
# plt.title('Stock Index Change')
# plt.plot(x_1, SI_CR,'g')

plt.figure(dpi = 75)
plt.title('Mapping Graph')
#plt.plot(x, GDP_map,'orange', label='GDP')
plt.plot(x, ER_map,'violet',label='Exchange Rate')
plt.plot(x, SI_map,'g',label='Stock Index')
plt.legend()

plt.show()
############################################# 그래프 #############################################
```
