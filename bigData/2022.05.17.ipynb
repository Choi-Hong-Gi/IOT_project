```python
import csv
import matplotlib.pyplot as plt

file = open('subwayfee.csv')
data = csv.reader(file)
next(data)

mx = [0] * 4
mx_station = [''] * 4
label = ['유임승차', '유임하차', '무임승차', '무임하차']
index = [''] * 4
color = ['#14ccc0','#95b674','#cd34f2','#cc1e2a']

for row in data:
    for i in range(4,8):
        row[i] = int(row[i])
        if row[i] > mx[i-4]:
            mx[i-4] = row[i]
            mx_station[i-4] = row[1] + ' ' + row[3]
            index[i-4] = row
            

for line in index:
    plt.figure(dpi = 100)
    plt.rc('font', family = 'Malgun Gothic')
    plt.title(line[1] +' '+ line[3])
    plt.pie(line[4:8], labels = label, autopct = '%.2f%%', colors = color)
    plt.axis('equal')
    #plt.savefig(line[1] +' '+ line[3] + '.png')
    plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

file = open('subwaytime.csv')
data = csv.reader(file)
next(data)
next(data)

s_in = [0] * 24
s_out = [0] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[i * 2 + 4]
        s_out[i] += row[i * 2 + 5]
        
plt.rc('font', family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4,28)) # x축 범위 설정
plt.show()
```
